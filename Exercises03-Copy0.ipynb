**Exercise 1 (10 pts)**. Avoding catastrophic cancellation.

The tail of the standard logistic distributon is given by $1 - F(t) = 1 - (1+e^{-t})^{-1}$.

- Define a function `f1` to calculate the tail probability of the logistic distribution using the formula given above
- Use `sympy` to find the exact value of the tail distribution (using the same symbolic formula) to 20 decimal digits
- Calculate the *relative error* of `f1` when $t = 25$ (The relative error is given by `abs(exact - approximate)/exact`)
- Rewrite the expression for the tail of the logistic distribution using simple algebra so that there is no risk of cancellation, and write a function `f2` using this formula. Calculate the *relative error* of `f2` when $t = 25$. 
- How much more accurate is `f2` compared with `f1` in terms of the relative error?

````python
# Your code here
# 1) Define a function f1 to calculate the tail probability
import math
def f1(t):
    return 1 - (1 + math.exp(-t))**(-1)

# 2) Use sympy to find the exact value of the tail distribution
import sympy
t = sympy.symbols('t')
taildist = 1 - (1 + sympy.exp(-t))**(-1)
# find the exact value to 20 decimal digits
exact = sympy.N(taildist.subs(t,25),20)
print "2) The exact value of the tail distribution is", exact   
    
# 3) Relative error 
approximatef1 = f1(25)
print "3) The relative error of f1 when t=25 is", abs(exact - approximatef1)/exact

# 4) Rewrite and calculate the relative error 
# Rewrite the expression for the tail and we get exp(-t)/(1+exp(-t))
def f2(t):
    return math.exp(-t)/(1 + math.exp(-t))
approximatef2 = f2(25)
print "4) The relative error of f2 when t=25 is", abs(exact - approximatef2)/exact

# 5) How much more accurate is f2 compared with f1 in terms of the relative error
print "5) f2 is", (abs(exact - approximatef1)/exact)/(abs(exact - approximatef2)/exact), "times accurate than f1"
````

- Write a formula that could solve the system of linear equations in terms of $X$ and $y$ Write a function `f1` that takes arguments $X$ and $y$ and returns $\beta$ using this formula.
- How could you code this formula using `np.linalg.solve` that does not require inverting a matrix? Write a function `f2` that takes arguments $X$ and $y$ and returns $\beta$ using this.
- Note that carefully designed algorithms *can* solve this ill-conditioned problem, which is why you should always use library functions for linear algebra rather than write your own.
```python
np.linalg.lstsq(x, y)[0]
```
- What happens if you try to solve for $\beta$ using `f1` or `f2`? Remove the column of $X$ that is making the matrix singluar and find the $p-1$ vector $b$ using `f2`.
- Note that the solution differs from that given by `np.linalg.lstsq`? This arises because the relevant condition number for `f2` is actually for the matrix $X^TX$ while the condition number of `lstsq` is for the matrix $X$. Why is the condition so high even after removing the column that makes the matrix singular?

````python
# Your code here
import numpy.linalg as la
# Load data
X = np.load('x.npy')
y = np.load('y.npy')
beta = np.load('b.npy')

# 1) Write a function f1
# the formula for beta is inv(X.T * X)* X.T * y
def f1(X, y):
    return np.dot(np.dot(la.inv(np.dot(X.T,X)),X.T), y)
    
# 2) Code this formula using np.linalg.solve that does not require inverting a matrix
# we can use SVD decomposition:
# X = SVD; X*beta = y; SVD*beta = y; S.T*S = I
# So VD*beta = S.T*y
def f2(X,y):
    U, s, V = np.linalg.svd(X, full_matrices=False)
    S = np.diag(s)
    return np.linalg.solve(np.dot(S,V), np.dot(U.T,y))
    
    
print "2) Using np.linalg.solve that does not require inverting a matrix, we get beta as",f2(X,y)    
   
# 3) Use library functions for linear algebra
print "3) Use library functions, we get beta as", np.linalg.lstsq(X, y)[0]    

# 4) Remove the column of X that is making the matrix singluar and find the p−1 vector b using f2.
print "4) Solving for beta using f1 gives us error."
print "   Solving for beta using f2 gives us results that differ much from library function results." 
# Find the rank of X
print "   The original rank of X is", la.matrix_rank(X)
# Remove the column of X that is making the matrix singluar
for i in range (0, la.matrix_rank(X)+1):
    newX = np.delete(X, i, 1)
    if la.matrix_rank(newX) == la.matrix_rank(X):
        break

print "   The p−1 vector beta using f2 is", f2(newX, y)
print "   The p−1 vector beta using library function is", np.linalg.lstsq(newX, y)[0] 

# 5) Why is the condition so high even after removing the column that makes the matrix singular?

U1, s1, V1 = np.linalg.svd(np.dot(newX.T,newX))
print(max(s1)/min(s1))

U2, s2, V2 = np.linalg.svd(np.dot(X.T,X))
print(max(s2)/min(s2))

U3, s3, V3 = np.linalg.svd(newX)
print(max(s3)/min(s3))

U4, s4, V4 = np.linalg.svd(X)
print(max(s4)/min(s4))

print newX
print "5) The condition number is high because the scale of column 2 in X is much larger than other columns"
````

````python
# Your code here
from IPython.display import Latex
from scipy.optimize import curve_fit
# 1) Implement bubble sort
A = [9,3,7,5,2,1]
print "1) Original list is", A
def bubble(list):
    for i in range(len(list)):
        for j in range( len( list ) - 1, i, -1 ):
            if ( list[j] < list[j - 1] ):
                swap = list[j]
                list[j] = list[j - 1]
                list[j - 1] = swap

bubble(A)               
print "   Bubble sorted list is", A
````

**Exercise 4 (20 pts)**. One of the goals of the course it that you will be able to implement novel algorihtms from the literature. 

- Implement the mean-shift algorithm in 1D as described [here](http://homepages.inf.ed.ac.uk/rbf/CVonline/LOCAL_COPIES/TUZEL1/MeanShift.pdf). 
    - Use the following function signature
    ```python
    def mean_shift(xs, x, kernel, max_iters=100, tol=1e-6):
    ```
    - xs is the data set, x is the starting location, and kernel is a kernel function
    - tol is the difference in $||x||$ across iterations
- Use the following kernels with bandwidth $h$ (a default value of 1.0 will work fine)
    - Flat - return 1 if $||x|| < h$ and 0 otherwise
    - Gaussian 
    $$\frac{1}{\sqrt{2 \pi h}}e^{\frac{-||x||^2}{h^2}}$$
    - Note that $||x||$ is the norm of the data point being evaluated minus the current value of $x$
- Use both kernels to find all 3 modes of the data set in `x1d.npy`
- Modify the algorithm and/or kernels so that it now works in an arbitrary number of dimensions.
- Use both kernels to find all 3 modes of the data set in `x2d.npy`
- Plot the path of successive intermediate solutions of the mean-shift algorithm starting from `x0 = (-4, 10)` until it converges onto a mode in the 2D data for each kernel. Superimpose the path on top of a contour plot of the data density.

````python
# First load and visualize data
import seaborn as sns
x1d = np.load('x1d.npy')
plt.hist(x1d, bins=20, color=sns.desaturate("indianred", .75))
plt.title("Histogram of x1d")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
````

````python
x2d = np.load('x2d.npy')
data = pd.DataFrame(x2d, columns=("X", "Y"))
sns.jointplot("X", "Y", data, kind="kde", color="blue")
plt.title("Plot of x2d")
````

````python
#  Define mean_shift function
def mean_shift(xs, x, kernel, max_iters=100, tol=1e-6):
    iterations = 0
    while iterations < max_iters:
        a = 0
        b = 0
        iterations = iterations + 1
        for i in range(len(xs)):
            norm = la.norm(x - xs[i])
            a = a + xs[i] * kernel(norm)
            b = b + kernel(norm)
        mhx = a / b - x
        converge = abs(mhx) < tol
        if converge:
            break
        else:
            x = x + mhx
    return x + mhx
        
# Define flat kernel function
def flat(x, h = 1.0):
    if abs(x) < h:
        return 1.0
    else:
        return 0.0

# Define Gaussian kernel function
def Gaussian(x, h = 1.0):
    return np.exp(-(x/h)**2)/np.sqrt(2*np.pi*h)  
````

````python
# Find all 3 modes of the data set in x1d.npy
flat1 = mean_shift(x1d, x1d[0], flat, max_iters=100, tol=1e-6)
flat2 = mean_shift(x1d, 1, flat, max_iters=100, tol=1e-6)
flat3 = mean_shift(x1d, x1d[-1], flat, max_iters=100, tol=1e-6)
print "Using Flat kernels to find all 3 modes in x1d.npy, we get", flat1, flat2, flat3
````

````python
gaussian1 = mean_shift(x1d, x1d[0], Gaussian, max_iters=100, tol=1e-6)
gaussian2 = mean_shift(x1d, x1d[len(x1d)/2], Gaussian, max_iters=100, tol=1e-6)
gaussian3 = mean_shift(x1d, x1d[-1], Gaussian, max_iters=100, tol=1e-6)
print "Using Gaussian kernels to find all 3 modes in x1d.npy, we get", gaussian1, gaussian2, gaussian3
````

````python
# Modify the algorithm and/or kernels so that it now works in an arbitrary number of dimensions.
# Define mean_shift function for 2d
def mean_shift_2d(xs, x, kernel, max_iters=100, tol=1e-6):
    iterations = 0
    while iterations < max_iters:
        a = 0
        b = 0
        iterations = iterations + 1
        for i in range(len(xs)):
            norm = la.norm(x - xs[i])
            a = a + xs[i] * kernel(norm)
            b = b + kernel(norm)
        mhx = a / b - x
        converge = abs(la.norm(mhx)) < tol
        if converge:
            break
        else:
            x = x + mhx
    return x + mhx
````

````python
# Plot the path of successive intermediate solutions of the mean-shift algorithm

# First rewrite the function so that it returns the list of path (in array)
def mean_shift_path(xs, x, kernel, max_iters=100, tol=1e-6):
    iterations = 0
    path = [x]
    while iterations < max_iters:
        a = 0
        b = 0
        for i in range(len(xs)):
            norm = la.norm(path[iterations] - xs[i])
            a = a + xs[i] * kernel(norm)
            b = b + kernel(norm)
        mhx = a / b - path[-1]
        converge = abs(la.norm(mhx)) < tol
        path.append(a / b)
        if converge:
            break
        iterations = iterations + 1
    return np.array(path)
````

````python
# Using flat kernel
print "As professor said, the starting point is not x0 = (-4, 10), instead we can define a different starting point"
lf = mean_shift_path(x2d, [5, 6], flat, max_iters=100, tol=1e-6)
print "The successive intermediate solutions (using flat kernel) is", lf
````
