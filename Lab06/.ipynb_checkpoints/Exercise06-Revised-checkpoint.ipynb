````python
# 1) What is the minimum distance she has to travel to do this?
# Can solve using scipy.optimize and constrained minimization
from scipy.optimize import minimize
cons = ({'type': 'eq', 'fun': lambda x: x[1]*(1 + x[0]) - 10})
def fun(x):
    return np.linalg.norm((2 - x[0], 8 - x[1])) + np.linalg.norm((x[0] - 8, x[1] - 4)) 

mini = minimize(fun, (1.0, 1.0), constraints = cons)
print "The minimum distance is", mini.fun
print "Corresponding point is", mini.x
````

````python
# 2) Modify the code below so that the path taken by the milkmaid is shown as a dashed line
x = np.linspace(0, 10, 100)
plt.xlim([0, 10])
plt.plot(x, 10.0 / (1 + x), 'b', linewidth = 5)
plt.scatter([2, 8, mini.x[0]], [8, 4, mini.x[1]], c = [0, 2, 1], s=100)
plt.plot([2, mini.x[0]], [8, mini.x[1]], 'r--')
plt.plot([mini.x[0], 8], [mini.x[1], 4], 'r--')
plt.text(8, 5, 'B (8,4)', fontsize=16)
plt.text(2, 9, 'A (2,8)', fontsize=16)
plt.text(0.5, 2, 'River', fontsize=16)
plt.text(0.5, 0.5, 'g(x y) = y - 10/(1+x)', fontsize=16);
````

**Exercise 2 (10 points)**.

You are given 5 samples of 100 rolls of a die. The die is chosen at random from one of two dice, either of which may be biased. Given the following counts, estimate the parameters for each die (i.e. the probabilites of getting a 1, 2, 3, 4, 5 or 6).

````python
df = pd.read_csv('dice.csv', index_col=0)
df
````

````python
data = np.array(df[0:5])
thetas = np.array([[0.05, 0.1, 0.15, 0.2, 0.2, 0.3], [0.2, 0.2, 0.2, 0.1, 0.1, 0.2]])
print data
print thetas
````

````python
def estimate(xs, thetas, max_iter=100, tol=1e-6):
    ll_old = -np.infty
    for i in range(max_iter):
        ll = np.array([np.sum(xs * np.log(theta), axis=1) for theta in thetas])    
        like = np.exp(ll)
        ws = like / like.sum(0)
        vs = np.array([w[:, None] * xs for w in ws])
        thetas = np.array([v.sum(0) / v.sum() for v in vs])
        ll_new = np.sum([w * l for w, l in zip(ws, ll)])
        if np.abs(ll_new - ll_old) < tol:
            break
        ll_old = ll_new
    return i, thetas, ll_new


i, thetas, ll = estimate(data, thetas)
print "The probaility of landing on each side for one die is", thetas[0]
print "The probaility of landing on each side for the other die is", thetas[1]
````

**Exercise 3 (10 points)**

You have some data about the intervals between arrivals (in minutes) of customers to a bank. You want to know how many customers to expect from 9 AM to 5 PM. Simulate 1000 sequences of customer arrivals over an 8 hour period and esitmate the mean and variance. 

````python
data = np.loadtxt('arrival_intervals.txt')
from scipy.interpolate import interp1d
# I studied how to do this from StackOverflow http://stackoverflow.com/questions/2745329/how-to-make-scipy-interpolate-give-an-extrapolated-result-beyond-the-input-range
def extrap1d(interpolator):
    xs = interpolator.x
    ys = interpolator.y
    def pointwise(x):
        if x < xs[0]:
            return ys[0] + (x - xs[0]) * (ys[1] - ys[0]) / (xs[1] - xs[0])
        elif x > xs[-1]:
            return ys[-1] + (x - xs[-1]) * (ys[-1] - ys[-2]) / (xs[-1] - xs[-2])
        else:
            return interpolator(x)
    def ufunclike(xs):
        return np.array(map(pointwise, np.array(xs)))
    return ufunclike

from statsmodels.distributions.empirical_distribution import ECDF
x = data
empirical_cdf = ECDF(x) 

# estimate inverse CDF function using interpolation
inv_cdf = extrap1d(interp1d(empirical_cdf.y, empirical_cdf.x, bounds_error = False, assume_sorted = True))

r = np.random.uniform(0, 1, 100)
ys = inv_cdf(r)
plt.hist(x, 25, histtype='step', color='red', normed=True, linewidth=1)
plt.hist(ys, 25, histtype='step', color='blue', normed=True, linewidth=1)
plt.show()
````

````python
# Simulate 1000 sequences of customer arrivals over an 8 hour period
# Generate 1000 sequence of random numbers
xb = np.array([inv_cdf(np.random.uniform(0,1,100)) for i in range(1000)])
# Store length of first n numbers that add up to 8 hours
length = [len(xb[i][np.cumsum(xb[i]) < 480]) for i in range(1000)]
# Esitmate the mean and variance
print "Mean of number of customers coming over an 8 hour period is", np.mean(length)
print "Variance of number of customers coming over an 8 hour period is", np.var(length)
````

**Exercsie 4 (10 points)**

The Cauchy distribution is given by 
$$
f(x) = \frac{1}{\pi (1 + x^2)}, \ \ -\infty \lt x \lt \infty 
$$

1. Integrate the tail probability $P(X > 2)$ using Monte Carlo with 1,000 samples:
    1. Sampling from the Cauchy distribution directly
    2. Sampling from the uniform distribution using an appropriate change of variables
2. Plot the 95% CI for the Monte Carlo estimates for the number of samples from 1 to 1,000, and using 1,000 reps to estimate the lower and upper bounds:
    1. For sampling from the Cauchy distribution using mulitple Monte Carlo sequences
    2. For sampling from the uniform distribution using bootstrap samples of a single Monte Carlo sequence



````python
# 1) Integrate the tail probability P(X>2)
# A. Sampling from the Cauchy distribution directly
n = 1000
x = stats.cauchy().rvs(n)
Sampling_Cauchy = 1.0 / n * np.sum(x > 2)
print "Sampling from the Cauchy distribution directly, we get P(X>2) =", Sampling_Cauchy
# B. Sampling from the uniform distribution using an appropriate change of variables
y = stats.uniform().rvs(n)
Sampling_Unif = 1.0 / n * np.sum(2.0 / (np.pi * (4 + y**2)))
print "Sampling from the uniform distribution using y = 2x, we get P(X>2) =", Sampling_Unif
````

````python
# 2) Plot the 95% CI and estimate the lower and upper bounds
# A. For sampling from the Cauchy distribution using mulitple Monte Carlo sequences
reps = 1000
xCauchy = np.random.standard_cauchy((n, reps))
yCauchy = 1 / np.arange(1, n + 1)[:, None] * np.cumsum(xCauchy, axis=0)
upperCauchy, lowerCauchy = np.percentile(yCauchy, [2.5, 97.5], axis = 1)
plt.plot(np.arange(1, n+1), upperCauchy, 'b', np.arange(1, n+1), lowerCauchy, 'b')
plt.xlabel('Sample Size')
plt.title('lower and upper bounds of 95% CI for sampling from the Cauchy distribution') 
plt.show()
````

````python
# B. For sampling from the uniform distribution using bootstrap samples of a single Monte Carlo sequence
xunif = np.random.choice(xCauchy[:, 0], (n, reps), replace = True)
yunif = 1 / np.arange(1, n+1)[:, None] * np.cumsum(xunif, axis=0)
upperunif, lowerunif = np.percentile(yunif, [2.5, 97.5], axis=1)
plt.plot(np.arange(1, n+1), upperunif, 'b', np.arange(1, n+1), lowerunif, 'b')
plt.xlabel('Sample Size')
plt.title('lower and upper bounds of 95% CI for sampling from uniform distribution') 
plt.show()
````

Estimate the following integral using Monte Carlo integration

\int_{-\infty}^{\infty} x^2 \frac{1}{2}e^{-|x|} dx

````python
# Using importance sampling
n = 10000
y = stats.norm(scale = 2).rvs(n)
h = 1.0 / n * np.sum(y**2 * stats.laplace.pdf(y) / stats.norm(scale = 2).pdf(y))

print "Estimateed integral using Monte Carlo integration is",h
````
